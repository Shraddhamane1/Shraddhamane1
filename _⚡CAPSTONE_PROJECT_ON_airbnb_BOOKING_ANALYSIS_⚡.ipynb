#⭐**TOP 5 EXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE**

# **⭐ TOP 5 AFFORDABLE NEIGHBOURHOOD BASED ON AVERAGE PRICE**

```python
temp = df.groupby(['neighbourhood'])['price'].median().reset_index()

temp = temp.sort_values(by = 'price', ascending = True)

top_5_affordable = temp.head(5)

print(top_5_affordable)
```

```python
# Chart - 7 visualization code

import matplotlib.pyplot as plt

# Assuming you have 'top_5_affordable' DataFrame with 'neighbourhood' and 'price' columns

# Data for the pie chart
neighborhoods = top_5_affordable['neighbourhood']
prices = top_5_affordable['price']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(prices, labels=neighborhoods, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightpink'])
plt.title('Top 5 Affordable Neighborhoods')

plt.show()
```
